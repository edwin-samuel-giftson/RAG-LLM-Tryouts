```python
# We start by fetching a document that loads the book - "Atomic habits"

from langchain_community.document_loaders import PyPDFLoader

pdfloader = PyPDFLoader("/content/drive/MyDrive/Colab Notebooks/RAG/RAG-data/Fostering_a_Positive_Workplace_Culture_Impacts_on_.pdf")
pdfpages = pdfloader.load_and_split()

len(pdfpages)
```

```python
mydocuments = pdfloader.load()
len(mydocuments)
```

```python
from langchain.text_splitter import CharacterTextSplitter
from langchain_community.vectorstores import FAISS
from langchain_openai.embeddings import OpenAIEmbeddings

text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
texts = text_splitter.split_documents(mydocuments)
```

```python
import os
# os. environ[ "OPENAI_API_KEY"] = [token]
```

```python
# Using OpenAl default embedding model
embeddings = OpenAIEmbeddings()

# Using LangChain API functions to interact with FAISS
db = FAISS.from_documents(texts, embeddings)
```

```python
db
```

```python
#Using Langchain retreiver

retriever = db.as_retriever()
from langchain.agents.agent_toolkits import create_retriever_tool
tool = create_retriever_tool(
    retriever,
    "Fostering_a_Positive_Workplace_Culture_Impact_Insights",
    "Searches and returns documents related to fostering a positive workplace culture and its impacts"
)
tools = [tool]
```

```python
#wrapping LLM with conversational agent

from langchain.agents.agent_toolkits import create_conversational_retrieval_agent

# LangChain given API to interact with chat models
from langchain_openai.chat_models import ChatOpenAI
llm = ChatOpenAI(temperature = 0)
agent_executor = create_conversational_retrieval_agent(llm, tools, verbose=True)
```

```python
llm.invoke("How does communication impact the workplace culture?")
```

```python
input = "What are the benefits of creating a positive workplace culture for an organization?"
result = agent_executor.invoke({"input": input})
```
