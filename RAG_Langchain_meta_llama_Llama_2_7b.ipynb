## RAG System Using Llama2 With Hugging Face

```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, ServiceContext
from llama_index.llms.huggingface import HuggingFaceLLM
from llama_index.core.prompts.prompts import SimpleInputPrompt
```

```python
documents=SimpleDirectoryReader("/content/drive/MyDrive/Colab Notebooks/RAG/RAG-data").load_data()
documents

#/content/drive/MyDrive/Colab Notebooks/RAG/RAG-data/Fostering_a_Positive_Workplace_Culture_Impacts_on_.pdf"
```

```python
len(documents)
```

```python
system_prompt="""
You are a Q&A assistant. Based on the provided context, respond to the question below while adhering to these guidelines:
1. If the answer is unclear or not found, do not speculate. Instead, state, "I do not know the answer"
2. If the answer is found, provide a clear response.
"""

## Default format supportable by LLama2
query_wrapper_prompt=SimpleInputPrompt("<|USER|>{query_str}<|ASSISTANT|>")
```

```python
!huggingface-cli login
```

```python
import torch

llm = HuggingFaceLLM(
    context_window=4096,
    max_new_tokens=256,
    generate_kwargs={"temperature": 0.0, "do_sample": False},
    system_prompt=system_prompt,
    query_wrapper_prompt=query_wrapper_prompt,
    tokenizer_name="meta-llama/Llama-2-7b-chat-hf",
    model_name="meta-llama/Llama-2-7b-chat-hf",
    device_map="auto",
    # uncomment this if using CUDA to reduce memory usage
    model_kwargs={"torch_dtype": torch.float16 , "load_in_8bit":True}
)
```

```python
# from langchain.embeddings.huggingface import HuggingFaceEmbeddings
from langchain_huggingface import HuggingFaceEmbeddings
from llama_index.embeddings.langchain import LangchainEmbedding

# Force CPU or GPU depending on your preference
embed_model = LangchainEmbedding(
    HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2", model_kwargs={"device": "cpu"})
)
```

```python
from llama_index.core import ServiceContext, set_global_service_context
from llama_index.core.node_parser import SentenceSplitter

service_context=ServiceContext.from_defaults(
    chunk_size=1024,
    llm=llm,
    embed_model=embed_model,
)
set_global_service_context(service_context)
```

```python
service_context
```

```python
index=VectorStoreIndex.from_documents(documents,service_context=service_context)
```

```python
from llama_index.core import Settings

Settings.llm = llm
Settings.embed_model = embed_model
Settings.chunk_size = 1024
```

```python
index = VectorStoreIndex.from_documents(documents, settings=Settings)
```

```python
index
```

```python
query_engine = index.as_query_engine()
```

```python
response = query_engine.query("What are the benefits of creating a positive workplace culture for an organization?")
```

```python
print(response)
```
